```python
def plotcircles():
    for i in range(len(radii)):
        cc = plt.Circle((centres[i].real,centres[i].imag), radii[i] ,fill=False, color=np.random.rand(3,),lw=1)
        ax.add_artist( cc )
        #plt.text(centres[i].real,centres[i].imag,k[i],size='xx-small')
        #ax.annotate(i,xy=(centres[i].real,centres[i].imag))
```

```python
from itertools import combinations
def nextcircle(k1,z1,k2,z2,k3,z3,n):
    if n>=10:
        return
    
    global centres,radii,k
    
    if len(radii)>=100000:
        return
    k4_1=k1+k2+k3+2*np.sqrt(k1*k2+k2*k3+k3*k1)
    k4_2=k1+k2+k3-2*np.sqrt(k1*k2+k2*k3+k3*k1)
    z4_1=(z1*k1+z2*k2+z3*k3+2*np.sqrt(k1*z1*k2*z2+k2*z2*k3*z3+k3*z3*k1*z1))/k4_1
    z4_2=(z1*k1+z2*k2+z3*k3-2*np.sqrt(k1*z1*k2*z2+k2*z2*k3*z3+k3*z3*k1*z1))/k4_2
    
    if abs(z4_1)>1 or abs(z4_2)>1:
        z4_1=z4_1*k4_1/k4_2
        z4_2=z4_2*k4_2/k4_1
        temp=k4_1
        k4_1=k4_2
        k4_2=temp 
    kz=[(k1,z1),(k2,z2),(k3,z3)]
    
    z4_1dist=np.absolute(centres-z4_1)
    z4_2dist=np.absolute(centres-z4_2)
    """
    if len(radii)==72:
        print(kz,k4_1,k4_2,z4_1,z4_2)
        print(np.min(z4_1dist[1:]-radii[1:]))"""
        
    if np.min(z4_1dist)>10**(-8) and np.min(z4_2dist)>10**(-8) and len(radii)>=10:
        z4_1=z4_1*k4_1/k4_2
        z4_2=z4_2*k4_2/k4_1
        temp=k4_1
        k4_1=k4_2
        k4_2=temp
        z4_1dist=np.absolute(centres-z4_1)
        z4_2dist=np.absolute(centres-z4_2)
        
    """if np.min([1/k4_1,1/k4_2])<10**(-8):
        return"""
    
    if np.min(z4_1dist)>10**(-8):
        radii=np.append(radii,1/k4_1)
        k=np.append(k,k4_1)
        centres=np.append(centres,z4_1)
        for kz1,kz2 in combinations(kz,2):
            nextcircle(*kz1,*kz2,k4_1,z4_1,n+1)
    

    #z4_2dist=np.absolute(centres-z4_2)
    if np.min(z4_2dist)>10**(-8):
        radii=np.append(radii,1/k4_2)
        k=np.append(k,k4_2)
        centres=np.append(centres,z4_2)
        for kz1,kz2 in combinations(kz,2):
            nextcircle(*kz1,*kz2,k4_2,z4_2,n+1)
    
    stdout.write("\r level: {} circles: {}".format(n,len(radii)))
    stdout.flush()
```

```python
r1=1
k1=-1
z1=0+0j
r2=np.random.rand()
r2=0.5
k2=1/r2
z2=(1-r2)+0j
r3=1-r2
k3=1/r3
z3=(-1+r3)+0j
radii=np.array([])
centres=np.array([])
radii=np.append(radii,[r1,r2,r3])
centres=np.append(centres,[z1,z2,z3])
k=1/radii
k[0]=-1.0
```

```python
n=0
nextcircle(k1,z1,k2,z2,k3,z3,n)
```
